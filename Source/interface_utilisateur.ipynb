```python
import streamlit as st
import pandas as pd

from engine_controller import Controller



# Initialisation du Controller
controller = Controller()

print(controller.csv_path) 

# Chargement des DataFrames locaux
tickers_df = pd.DataFrame({
    "Ticker": ["AAPL UW", "1COV GY"],
    "Country": ["US", "GE"],
})
data_sources = {"tickers_df": tickers_df}

# UI avec des listes déroulantes
st.title("Indice Finance - Création d'Indices")
ticker = st.selectbox("Sélectionnez un Ticker", options=tickers_df["Ticker"].unique())
country = st.selectbox("Sélectionnez un Pays", options=tickers_df["Country"].unique())

# Bouton pour exécuter le traitement
if st.button("Générer l'Indice"):
    # Mise à jour des paramètres
    controller.set_parameters(ticker, country)
    
    # Extraction des données
    filtered_data = controller.fetch_data(data_sources)
    
    # Exécution du modèle
    result = controller.process_model(filtered_data)
    
    # Affichage des résultats
    st.write("Résultat du modèle :")
    st.json(result)
```
